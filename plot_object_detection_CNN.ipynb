```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca file CSV
df = pd.read_csv('lintasan_surma_kiri.csv')

# Ganti 'kolom1' dan 'kolom2' dengan nama kolom yang sesuai dari file Anda
plt.scatter(df['Time'], df['Steering'])

plt.title('Grafik Scatter')
plt.xlabel('Time')
plt.ylabel('Steering')
plt.show()
```
